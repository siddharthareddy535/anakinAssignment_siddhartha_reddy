```python
import pandas as pd
import requests
import urllib3
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(options=options)
driver.get("https://www.bigbasket.com/")
```

```python
category=[]
sub_category=[]
sub_sub_category=[]
brand_slug=[]
brand_name=[]
brand_id=[]
headers=requests.utils.default_headers()
headers.update({'User-agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64)"})
r=requests.get("https://www.bigbasket.com/auth/get_menu/?city_id=1",headers=headers)
page=json.loads(r.content)
x=page['topcats']

# x[0]['top_category']['brands']

for i in range(len(page['topcats'])):
    length=len(x[i]['top_category']['brands'])
    p=0
    for j in range(len(x[i]['sub_cats'])):
        for k in range(len(x[i]['sub_cats'][j])):
            for l in range(len(x[i]['sub_cats'][j][k]['cats'])):
                category.append(x[i]['top_category']['name'])
                sub_category.append(x[i]['sub_cats'][j][k]['sub_category'][0])
                sub_sub_category.append(x[i]['sub_cats'][j][k]['cats'][l]['cat'][0])
                brand_slug.append(x[i]['top_category']['brands'][p]['brand_slug'])
                brand_name.append(x[i]['top_category']['brands'][p]['brand_name'])
                brand_id.append(x[i]['top_category']['brands'][p]['brand_id'])
            p=p+1
            if p>=length:
                p=0

            
df1=pd.DataFrame(category)
df2=pd.DataFrame(sub_category)
df3=pd.DataFrame(sub_sub_category)
df4=pd.DataFrame(brand_slug)
df5=pd.DataFrame(brand_name)
df6=pd.DataFrame(brand_id)
final=pd.concat([df1,df2,df3,df4,df5,df6],axis=1)
columns=['category','sub-category','sub-sub-category','brand_slug','brand_name','brand_id']
final.columns=columns
pd.set_option('display.max_rows', None)
final
```
