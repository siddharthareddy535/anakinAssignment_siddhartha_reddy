```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(options=options)
# getting acess of site 
driver.get("https://food.grab.com/ph/en/restaurants") 
```

```python
lists=[]
z=[]
# extracting  required data from class
x=driver.find_element(By.CLASS_NAME,"ant-layout").get_attribute("innerHTML")
soup=BeautifulSoup(x,"html.parser")
for i in soup.findAll('div',{'class':'ant-col-lg-6'}):
    lists.append("https://food.grab.com/"+i.find('a').attrs['href'])
for i in lists:
    z.append(i.split("/").pop())
```

```python
answer=[]
for i in range(5):
    page = driver.find_element("id", "__NEXT_DATA__").get_attribute("innerHTML")
    # showing data in json form
    data = json.loads(page)
    # extracting content which contains names,latitude and longitude
    total_restaurents= data["props"]["initialReduxState"]["pageRestaurantsV2"][
        "entities"
]["restaurantList"]
    # finally pushing in list
    for i in total_restaurents:
        rest={}
        rest['name']=total_restaurents[i]['name']
        rest['latitude']=total_restaurents[i]['latitude']
        rest['longitude']=total_restaurents[i]['longitude']
        answer.append(rest)
    # for loading more restaurents
    load_more=driver.find_element(
       "xpath",'//*[@id="page-content"]/div[5]/div/div/div[4]/div/div/div'
    )
    load_more.click()
    sleep(10)
    
print(answer)
```

```python
# converting into dataframes
df1=pd.DataFrame(lists,columns=['url'])
df2=pd.DataFrame(z,columns=['id'])
df3=pd.DataFrame(answer,columns=['name','latitude','longitude'])
final=pd.concat([df2,df1],axis=1)
final
```
